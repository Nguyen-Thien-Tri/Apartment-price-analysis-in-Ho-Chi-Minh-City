```python
def datarame_with_multiple_conditions(
    df: pd.DataFrame, vars: list, values: list
) -> pd.DataFrame:
    conditions = list(zip(vars, values))

    temp_df = df.copy()
    for condition in conditions:

        temp_df = temp_df[temp_df[condition[0]] == condition[1]]
    return temp_df
```

```python
def morton_encode(lat, lon):
    # Số bit sử dụng cho mỗi latitude và longitude (cần điều chỉnh tùy thuộc vào độ chính xác cần thiết)
    BITS = 32

    def shift_bits(n):
        # Di chuyển bit của số nguyên n một vị trí sang phải
        n = (n | (n << 16)) & 0x0000FFFF0000FFFF
        n = (n | (n << 8)) & 0x00FF00FF00FF00FF
        n = (n | (n << 4)) & 0x0F0F0F0F0F0F0F0F
        n = (n | (n << 2)) & 0x3333333333333333
        n = (n | (n << 1)) & 0x5555555555555555
        return n

    # Chuyển đổi latitude và longitude thành số nguyên và cắt bớt phần thập phân
    lat_int = int(lat * (1 << BITS))
    lon_int = int(lon * (1 << BITS))

    # Mã hóa Morton từ latitude và longitude
    morton_code = shift_bits(lat_int) | (shift_bits(lon_int) << 1)
    return morton_code
```

```python
# Tạo cột mới là mã Morton của latitude và longitude
df["morton_code"] = df.apply(
    lambda row: morton_encode(row["ad_latitude"], row["ad_longitude"]), axis=1
)
# Correlation cho thấy không có sự tương quan giữa giá và mã Morton
df[["ad_price", "morton_code"]].corr()
```

```python
import seaborn as sns
from scipy.stats import gaussian_kde


def check_correlation_for_each_value(categorical_variables: list, df: pd.DataFrame):
    fig, axs = plt.subplots(3, 8, figsize=(16, 16))
    not_correlated_values_dict = {}
    i = 0
    for col in categorical_variables:
        print(col)
        not_correlated_values = []
        for value in df[col].unique():
            f, p = stats.f_oneway(
                df[df[col] == value]["ad_price"], df[df[col] != value]["ad_price"]
            )

            if p < 0.05:
                not_correlated_values.append(value)
                print(col, value)
                data = df[df[col] == value]["ad_price"]
                counts, bins, _ = axs.flat[i].hist(
                    data, bins=30, alpha=0.5, label=str(value), density=True
                )
                kde = gaussian_kde(data)
                kde_vals = kde(bins)
                axs.flat[i].plot(bins, kde_vals)
                axs.flat[i].legend()
                axs.flat[i].set_title(col)
                print("-")
                i += 1
        not_correlated_values_dict[col] = not_correlated_values
    plt.tight_layout()
    plt.show()
    return not_correlated_values_dict
```

```python
def find_best_value_for_each_categorical_variable(
    feature: str, target: str, df: pd.DataFrame
):
    unique_values = df[feature].unique()

    result = {}

    for val in unique_values:

        f, p = stats.f_oneway(
            df[df[feature] == val][target], df[df[feature] != val][target]
        )

        if p < 0.05:
            result[val] = p

    return min(result, key=result.get)
```

# Tìm kiểu căn hộ có giá tốt nhất


```python
from scipy.stats import f_oneway



def find_least_correlated_combo(df: pd.DataFrame, features: list, target: str):

    grouped = df.groupby(features, observed=True)[target]

    grouped = grouped.groups.items()

    grouped = dict(grouped)

    # chuyển các value của dict grouped thành list

    for feature in grouped:
        grouped[feature] = list(grouped[feature])

    result = {}

    for feature in grouped:

        temp_list = [grouped[k] for k in grouped.keys() if k != feature]

        # Chuyển temp_list thành một list các giá trị

        temp_list = [item for sublist in temp_list for item in sublist]

        temp = f_oneway(grouped[feature], temp_list)[1]

        if temp < 0.05:
            result[feature] = f_oneway(grouped[feature], temp_list)[0]

    if result == {}:
        return None

    least_correlated_features = min(result, key=result.get)
    return least_correlated_features
```

# Loại căn hộ nào ở quận nào cho giá thấp nhất


```python
# Xem từng loại căn hộ
types = df["ad_params_apartment_type_value"].unique()
```

```python
# Xem xét các quận
areas = df["ad_area_name"].unique()
```

```python
type_area_f = {}
for type in types:
    for area in areas:
        if (
            len(
                df[
                    (df["ad_params_apartment_type_value"] == type)
                    & (df["ad_area_name"] == area)
                ]
            )
            == 0
            or len(
                df[
                    (df["ad_params_apartment_type_value"] != type)
                    | (df["ad_area_name"] != area)
                ]
            )
            == 0
        ):
            continue
        f_oneway_result = stats.f_oneway(
            df[
                (df["ad_params_apartment_type_value"] == type)
                & (df["ad_area_name"] == area)
            ]["ad_price"],
            df[
                (df["ad_params_apartment_type_value"] != type)
                | (df["ad_area_name"] != area)
            ]["ad_price"],
        )
        if f_oneway_result[1] < 0.05:
            type_area_f[(type, area)] = f_oneway_result[0]
```
